### This script aims to compute dividend adjusted prices. 

```python
import numpy as np
import pandas as pd
```

```python
prices = pd.read_csv('NOK_prices.csv', header=0)
dividends = pd.read_csv('NOK_dividends.csv',header=0)
```

```python
# convert column 'date' from string to datetime64
prices['date'] = pd.to_datetime(prices.date, dayfirst=True)
dividends['date'] = pd.to_datetime(dividends.date, dayfirst=True)
```

```python
# obtain time interval we are interested in studying
start_date = '25-2-2010'  
end_date = '17-7-2015'
t_first = pd.to_datetime(start_date, dayfirst=True)
t_last = pd.to_datetime(end_date, dayfirst=True)
```

```python
# filter the data for the selected range
nok_stock = prices[(prices.date >= t_first) & (prices.date < t_last)]
nok_dividends = dividends[(dividends.date >= t_first) & (dividends.date< t_last)]
nok_stock = nok_stock.set_index('date')
nok_dividends = nok_dividends.set_index('date')
```

```python
# compute dividend adjusted prices
for t in nok_dividends.index:
    nok_stock.loc[t:, 'adj_value'] = nok_stock.loc[t:, 'close']*(1.0+nok_dividends.loc[t,'dividends']/nok_stock.loc[t, 'close'])
```

```python
nok_stock.loc[:nok_dividends.index[0],'adj_value'] = nok_stock.loc[:nok_dividends.index[0],'close']
```
